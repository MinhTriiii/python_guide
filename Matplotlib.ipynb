# **Matplotlib** in a nutshell (plots)

Create simple plots

```python
import matplotlib.pyplot as plt
import numpy as np 

x = np.linspace(0, 10, 100) # vector of length 100 with values from 0 to 10 
y = np.sin(x)  

plt.plot(x, y, label="legend", color="blue", linestyle="-")  
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.xlim((0,10))
plt.ylim((-1.1,1.1))
plt.grid()
plt.title("Title")
plt.legend()  
plt.show()
```

Examples for various charts

```python
# Bar Chart
labels = ["A", "B", "C", "D"]
values = [10, 20, 15, 30]

plt.bar(labels, values, color="skyblue")
plt.xlabel("groups")
plt.ylabel("values")
plt.title("Bar Chart")
plt.show()

# Histogram
data = np.random.randn(1000)  
plt.hist(data, bins=30, color="purple", edgecolor="black")
plt.title("Histogram")
plt.show()

# Scatterplot
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, color="green", alpha=0.5)  # alpha makes the dots transparent 
plt.title("Scatterplot")
plt.show()

# Boxplot
data = [np.random.normal(0, std, 100) for std in range(1, 5)]

plt.boxplot(data, vert=True, patch_artist=True, notch=False, labels=["group 1", "group 2", "group 3", "group 4"])
plt.title("Boxplot")
plt.xlabel("groups")
plt.ylabel("values")
plt.show()
```

Multiple Plots (subplots)

```python
fig, axs = plt.subplots(2, 2)  # 2x2 grid

axs[0, 0].plot(x, np.sin(x))
axs[0, 0].set_title("Sine")

axs[0, 1].plot(x, np.cos(x))
axs[0, 1].set_title("Cosine")

axs[1, 0].bar(["A", "B", "C"], [5, 10, 15])
axs[1, 0].set_title("Bar Chart")

axs[1, 1].hist(np.random.randn(1000), bins=30)
axs[1, 1].set_title("Histogram")

plt.tight_layout()  # to avoid overlapping
plt.show()
```

Nice to know

```python
# 3D plot
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

ax.plot_surface(X, Y, Z, cmap="viridis")

plt.show()

# saving as png or pdf
plt.savefig("my_plot.png", dpi=300)  # save as PNG (good quality)
plt.savefig("my_plot.pdf")  # save as PDF
```
