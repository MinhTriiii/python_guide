# **Git** in a nutshell (version control) 
## Git Basics

- Git is a version control system that helps you track changes in your code and manage different versions of your projects
- It allows you to save snapshots of your work through commits, so you can review, restore, or compare previous versions
- Git makes it easier to work on projects with multiple people by allowing developers to collaborate and merge their changes
- A repository (repo) is a project folder that is tracked by Git and contains the history of all changes
- Git allows you to create branches to develop new features or test changes without affecting the main codebase
- Changes can be shared with others using remote repositories such as GitHub, GitLab, or Bitbucket
- Git helps keep track of who changed what and when, which is especially useful for larger projects
- It provides tools to compare changes, resolve conflicts, and combine different versions of a project
- Git is independent of Python and can be used with almost any programming language or type of project

## Create or Clone a project (Connect to GitHub)
- (Creating) Go on GitHub and create new repo -> (in local project) git remote add origin <Git_URL> -> git push -u origin master
- (To *delete* the repo go on GitHub -> settings -> danger zone -> delete)
- (Cloning) git clone <Git_URL>

## Branches (test your own ideas without bothering other people)
- Show all branches: git branch [-a shows also remote branches] (*branch is active branch)
- Creating and switching to a new branch: git switch -c <new_branch>
- Switching to another branch: git switch <new_branch>
- Delete a local branch: git branch -d <new_branch> (deletes only if merged [-D forces it])
- Delete a remote branch: git push origin --delete <new_branch>

## Updating your branch with new content from master (merge & rebase)
```
master: A --- B --- E              master: A --- B --- E --- F             master: A --- B --- E --- F ------- I
               \            ->                    \           \     ->                    \           \       /
new_branch:     C --- D            new_branch:     C --- D --- G           new_branch:     C --- D --- G --- H 
```

If you work alone on your branch, then rebase might be worth. In case you don't know what you are doing just use merge

1) Merge (from master to <new_branch>)
- git switch <new_branch>
- git fetch origin
- git merge master

\+ complete history remains (you can see exactly when what happened)    
\- might become very messy with a large number of merges

```
master: A --- B --- E              master: A --- B --- E 
               \            ->                          \ 
new_branch:     C --- D            new_branch:           C' --- D' (C' and C are different commits) 
```

2) Rebase (put <new_branch> on master)
- git switch <new_branch>
- git fetch origin
- git rebase master

\+ linear history    
\- might change commits from other people on this branch

## Usual workflow 
- git switch main
- (common) git pull
- (see all changes) git fetch origin -> git log --oneline HEAD..origin/main -> git rebase origin/main
- ... *write new feature* ...
- git add <new_feature.ipynb>
- git commit -m "<(standardized)_message>"
- (initial push) git push -u origin <branch_name>
- (after initial) git push  

## Nice to know 
- You can't break anything if it's local and not pushed yet ¯\\_( ͡° ͜ʖ ͡°)_/¯
- **Always** double check if you want to delete anything
- When working in a team, don't just git merge <new_branch>. Use pull requests
- Sometimes use 'Squash and merge'. A lot of small commits are bundled to one commit and then merged.
- git status (shows current state)
- git add . (adds everything) 
- git log --oneline (compact commit-history)
- git stash / git stash pop (if your changes aren't good enough for a commit so you put them away for the moment)
- git remote -v (shows GitHub connection)
