# **Numpy** in a nutshell (working with numbers)

Creating a `ndarray` and some operations 

```python
import numpy as np

# vector
input_list = [1,2,3,4,5,6]
a = np.array(input_list) # or a = np.array([1,2,3,4,5,6])  
a_shape = np.shape(a)
print(a)
print(f"The vector a has the shape: {a_shape}")

# Better use reshape to get (3,1) ndarray
better_a = a.reshape((6,1))
matrix_a = a.reshape((3,2))
print(better_a)
print(f"The vector better_a has the shape: {better_a.shape}")
print(matrix_a)
print(f"The vector matrix_a has the shape: {matrix_a.shape}")

# matrix
A = np.array([[1,2,3],[4,5,6]])   
A_shape = np.shape(A)
A_flat = A.flatten()
print(A)
print(f"The matrix A has the shape: {A_shape}")
print(f"The matrix A' has the shape: {A.T.shape}")

# Usual way of initializing a matrix
np.zeros((3,4)) # 3x4 zero matrix
np.ones((3,4)) # 3x4 one matrix
np.full((3,4), 5) # 3x4 matrix filled with 5
np.eye(5) # 5x5 identity matrix
np.random.uniform(0,1,10) # 10 uniformly distributed variables in (0,1) as 1D array 
np.random.rand(3,4) # 3x4 matrix with random values in (0,1)
np.random.randint(1,9,(3,4)) # 3x4 matrix with random integers between 1 and 8
np.random.normal(0,1,(3,4)) # 3x4 matrix with random normal distributed numbers N(0,1)
```

Mathematical Operations

```python
# Reminder: a = [1,2,3,4,5,6].T

# elementwise operations
print(a + 1)
print(a * 2)
print(a ** 2)
print(np.exp(a))

# some statistical functions
print(np.mean(a))  
print(np.median(a))  
print(np.std(a))  
print(np.sum(a))  
print(np.min(a), np.max(a))

# matrix operations
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
print(A @ B)  
print(np.dot(A, B)) # alternative
print(A + B)
print(np.vstack((A, B)))
print(np.hstack((A, B)))

# linear algebra
print(np.linalg.inv(A))
print(np.linalg.det(A))
print(np.linalg.eig(A))
f = [2,3]
print(np.linalg.solve(A,f))
```
