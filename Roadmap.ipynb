# <u> **1. Basics in python** </u>
### <u> Standard libraries </u>
```python
import math
```    
- **Working with real numbers**: Do basic mathematics here, like *sin*, *log* or *sqrt*. For complex numbers use `cmath`

```python
import random
```    
- **Working with random variables**: For discrete and continuous random variables. **Not** actual random variables, only pseudo random. Set seeds for important scientific projects. 

```python
import time
```    
- **Tracking time in python for optimization**: Different ways to track runtime. For simple time tracking `time.time()`, for more precision `time.perf_counter()` or `time.process_time()`
- **Working with actual time**
- **Also see**: `datetime` more modern and object oriented, `calendar` to work with calendars, `sched` for scheduling
```python
import re
```    
- **Language to search, find and replace patterns in strings**: pattern matching (e.g. E-Mail, phone number, words), validating inputs, text processing (e.g. replace, split) and much more
```python
import os
```    
- **Operating system functions in python**: work with files, folders and processes independent of your operating system
- **Also see**: `shutil` (shell utilities) the big brother of `os` for functions that go beyond *creating* and *deleting* files, like *copying* or *moving*

### <u> Modular programming </u>
When working on a large project consider using modules (extra programs beside the main.py with functions, classes etc.). <br>
To do so use 
```python
import custom_module as shortcut
from custom_module import function_a, function_b  
```

# <u> **2. NumPy** </u>
```python
import numpy as np
```     
- **Working with Arrays**: In NumPy, the `ndarray`-object is the central element
- **Mathematical and Statistical Operations**: NumPy can perform various calculations such as sums `sum`, mean `mean`, 
standard deviation `std`, sine `sin`, logarithm`log` etc.
- **Linear Algebra**: Solving linear equation systems `linalg.solve`, inverse of a matrix `linalg.inv`, 
eigenvalue calculation `linalg.eig`, determinants `linalg.det`, various matrix operations â†’ addition A + B, multiplication `dot`, element-wise multiplication A * B, transposition `.T`
- **Nice to know**: 
    - NumPy is the foundation for many other libraries like Pandas, SciPy, TensorFlow, and Scikit-Learn
    - NumPy is implemented in C â†’ fast, efficient, and requires less memory
---
# <u> **3. Matplotlib** </u>
```python
import matplotlib.pyplot as plt
``` 
Matplotlib for creating **charts** and **graphs**. `pyplot` makes creating plots in Matplotlib significantly easier and is the standard approach for visualizations in Python.

---
# <u> **4. Pandas** </u>
# <u> **5. Scipy** </u>
# <u> **6. Scikit-Optimize** </u>
# <u> **7. Seaborn** </u>
# <u> **8. Polars** </u>
# <u> **9. Statsmodel** </u>
# <u> **10. Scikit-Learn** </u>
# <u> **11. Tensorflow** </u>
# <u> **12. SQL in Python und VSC** </u>
